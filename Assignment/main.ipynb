```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Here we load our csv file (Dataset)

```python
df = pd.read_csv("data.csv")
df
```

Here we take info for know whic column have str,int or float DataType

```python
df.info()
```

```python
df.describe()
```

This give us knowledge about how many null values in each column

```python
df.isnull().sum()
```

This gives us all unique values from 'make' column

```python
df['Make'].unique()
```

We update all column names in one format to simplyfy further cleaning

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
df
```

This is for delete all duplicate columns

```python
df = df.drop_duplicates()
print(f"Rows Ramaining: {len(df)}")
```

This is boxplot for 'engine_hp' that give us idea about outliers.

```python
sns.boxenplot(x=df['engine_hp'])
plt.show()
```

Here we fill our null values for each column with diffrent strategy

```python
median_hp = df.groupby('make')['engine_hp'].transform('median')
```

```python
df['engine_hp'] = df['engine_hp'].fillna(median_hp)
```

```python
# now still 18 values is null in engine_hp so we fill with global median

global_hp_median = df['engine_hp'].median()
```

```python
df['engine_hp'] = df['engine_hp'].fillna(global_hp_median)
```

```python
median_cly = df.groupby('make')['engine_cylinders'].transform('median')
```

```python
df['engine_cylinders'] = df['engine_cylinders'].fillna(median_cly)
```

```python
mode_door = df['number_of_doors'].mode()[0]
```

```python
df['number_of_doors'] = df['number_of_doors'].fillna(mode_door)
```

```python
df['market_category'] = df['market_category'].fillna("Unknown")
```

Here we want to know whic value is most frequent in 'engine_fuel_type' so we can fill null values with them

```python
df['engine_fuel_type'].value_counts()
```

```python
df['engine_fuel_type'] = df['engine_fuel_type'].fillna("regular unleaded")
```

Now we can see that there is no null values are left.

```python
df.isnull().sum()
```

```python
df.info()
```

# Here is the top 10 expensive car and bugatti is kind of outlier because its price is very high compare to others

```python
plt.Figure(figsize=(12,6))
df.groupby('make')['msrp'].median().sort_values(ascending=False).head(10).plot(kind='bar')
plt.ylabel("Median Price")
plt.title("Top 10 Most Expensive Car Brand")
plt.show()
```

# Here price distribution is right skewd means there is outliers of very high price and we cleary saw that there is many low price cars and very few high price cars


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['msrp'], bins=50, kde=True,color='purple')
plt.title("Distribution of Car Prices (MSRP)")
plt.xlim(0,200000)
plt.show()
```

# Here we can see that engine_hp have strong rellation with msrp

```python
num_cols = ['year', 'engine_hp', 'engine_cylinders', 'number_of_doors', 'highway_mpg', 'city_mpg', 'popularity', 'msrp']
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='RdBu_r')
plt.title("Numerical Feature Correlation")
plt.show()
```

# Here we can see that when engine_hp is increasing price is also increasing but for luxury cars price is increasing very high 

```python
plt.Figure(figsize=(10,6))
sns.scatterplot(x='engine_hp', y='msrp', data=df)

plt.title('Relationship: Horsepower VS Price')
plt.xlabel('Engine Horsepower (HP)')
plt.ylabel('Price (MSRP)')

sns.regplot(x='engine_hp', y='msrp', data=df, scatter=False, color='red')

plt.show()
```
